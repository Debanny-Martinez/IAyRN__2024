```python
#Librerias
import matplotlib.pyplot as plt #generar graficos
import numpy as np #generar arreglos
import pandas as pd #cargar conjunto de datos externos
import sklearn.linear_model #contiene diferentes modelos de ML
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Cargar conjunto de datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
#Dividir datos en conjunto de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2)
```

```python
#Crear el modelo de Regresión Lineal
modelo = LinearRegression()
```

```python
#Entrenar el modelo con los datos de entrenamiento
modelo.fit(train_data, train_labels);
```

```python
#Hacer predicciones
predicciones = modelo.predict(test_data)
```

```python
#Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predicciones)
print("Error Cuadrático Medio: ", mse)
```

```python
#Gráficar el modelo de regresión lineal con las predicciones
plt.scatter(test_data, test_labels, color='violet')
plt.plot(test_data, predicciones, color='lightblue')
plt.title('Regresión Lineal: Horas de Estudio - Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.show()
```
